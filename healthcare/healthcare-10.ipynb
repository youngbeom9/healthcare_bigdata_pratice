```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_ratio = pd.read_csv(r"C:\Users\302-28\Downloads\stacked_ratio.csv", encoding='CP949')
df_ratio.head(10)

plt.rc('font', family='Malgun Gothic')
plt.rc("figure", figsize=(10, 8))

ax = df_ratio.plot.bar(stacked=True)
ax.set_xticklabels(df_ratio['deviceId'])

handles, labels = ax.get_legend_handles_labels()

ax.legend(reversed(handles), reversed(labels), loc='best')
```

```python
df_ratio.head()
df_ratio.info()
```

부재 특성 제외 누적그래프

```python
plt.rc('font', family='Malgun Gothic')
plt.rc('figure',figsize=(10,8))

deviceId = df_ratio['deviceId']
fall = df_ratio['낙상']
caution = df_ratio['주의']
action = df_ratio['활동']
non_action = df_ratio['미활동']

plt.bar(deviceId,action,label ='활동')
plt.bar(deviceId,non_action, bottom=action,label='미활동')
plt.bar(deviceId,caution,bottom=action+non_action,label='주의')
plt.bar(deviceId,fall,bottom=action+non_action+caution,label='낙상')
plt.legend()
plt.show()
```

25번 기간별 활동

```python
df_ratio_25=pd.read_csv(r"C:\Users\302-28\Downloads\stacked_one_25.csv",dtype={'day':object, '부재':int, '미활동':int,
                              '활동':int,'주의':int,'낙상':int}, encoding="CP949")

plt.rc('figure',figsize=(15,9))
plt.rc('font',family='Malgun Gothic')

day=df_ratio_25['day']
absent = df_ratio_25['부재']
non_action = df_ratio_25['미활동']
action = df_ratio_25['활동']
caution = df_ratio_25['주의']
fall = df_ratio_25['낙상']

plt.bar(day,non_action,label='미활동')
plt.bar(day,action,bottom=non_action,label='활동')
plt.bar(day,absent,bottom=action+non_action,label='부재')
plt.bar(day,caution,bottom=absent+action+non_action,label='주의')
plt.bar(day,fall,bottom=absent+action+non_action+caution,label='낙상')

plt.legend()
plt.xticks(rotation=45) #x축 라벨 45도 회전
plt.show()
```
